Dilation, Erosion

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def erosion_dilation():
    img = cv2.imread('/content/image-photoaidcom-greyscale.jpg')
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    height, width = gray.shape

    kernel = np.array([[1, 1, 1],
                       [1, 1, 1],
                       [1, 1, 1]], dtype=np.uint8)

    erosion_cv2 = cv2.erode(gray, kernel)
    dilation_cv2 = cv2.dilate(gray, kernel)

    images = [erosion_cv2, gray, dilation_cv2]
    titles = ['OpenCV Erosion', 'Original', 'OpenCV Dilation']

    plt.figure(figsize=(18, 4))

    for i in range(len(images)):
        plt.subplot(1, 5, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.tight_layout()
    plt.show()


erosion_dilation()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def erosion_dilation():
    img = cv2.imread('/content/image-photoaidcom-greyscale.jpg')
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    height, width = gray.shape

    out = np.zeros((height + 2, width + 2), dtype=np.uint8)
    out[1:1 + height, 1:1 + width] = gray.copy()

    erosion = np.zeros_like(gray)
    dilation = np.zeros_like(gray)

    for i in range(height):
        for j in range(width):
            temp = out[i:i+3, j:j+3]
            erosion[i, j] = np.min(temp)
            dilation[i, j] = np.max(temp)

    kernel = np.array([[1, 1, 1],
                       [1, 1, 1],
                       [1, 1, 1]], dtype=np.uint8)


    images = [erosion, gray, dilation]
    titles = ['Manual Erosion', 'Original', 'Manual Dilation']

    plt.figure(figsize=(18, 4))

    for i in range(len(images)):
        plt.subplot(1, 5, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.tight_layout()
    plt.show()


erosion_dilation()
```
